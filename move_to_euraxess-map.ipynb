```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import googlemaps
import json
import folium
```

```python
load = requests.get("https://euraxess.ec.europa.eu/information/centres/search?page=0")
load_soup = BeautifulSoup(load.text,'html.parser')

centres_data = []
for page in range(0,54):
    load = requests.get("https://euraxess.ec.europa.eu/information/centres/search?page=" + str(page))
    load_soup = BeautifulSoup(load.text,'html.parser')


    centres = load_soup.find_all("div", {"class" : "views-row"})
    for i in centres:
        centre_datapoint = {}
        centre_title = i.find_all("h2")[0].text
        centre_location = i.find_all("ul", {"class" : "list-items"})[1].find_all("div", {"class" : "value"})[1].text
        centre_contact = i.find_all("ul", {"class" : "list-items"})[1].find_all("div", {"class" : "value"})[0].text
        centre_coverage = i.find_all("ul", {"class" : "list-items"})[0].find_all("div", {"class" : "value"})[0].text
        centre_type = i.find_all("ul", {"class" : "list-items"})[0].find_all("div", {"class" : "value"})[1].text
        centre_datapoint = {"title" : centre_title, "location" : centre_location, "contact" : centre_contact, "coverage" : centre_coverage, "type" : centre_type}
        centres_data.append(centre_datapoint)
        print(i.find_all("ul", {"class" : "list-items"})[1].find_all("div", {"class" : "value"})[1].text)

pd.DataFrame(centres_data).to_excel("centres_output.xlsx")
```
